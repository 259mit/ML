### Import Libraries

```python
import numpy as np
import pandas as pd
from pandas import DataFrame as df
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Dataset

```python
canada = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx',
 sheet_name='Canada by Citizenship',
 skiprows=range(20),
 skipfooter=2)
```

```python
canada.head()
#spare df
canada1 = canada
```

### Rename columns

```python
canada.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Continent-Region'}, inplace=True)
canada.drop(['AREA','REG','DEV','Type','Coverage'], inplace=True, axis='columns')
canada.isnull().sum().sum()
```

```python
canada.head(5)
```

```python
canada.index.values
```

### Reassign Index

```python
canada.set_index('Country', inplace=True)
```

### Convert columns to str

```python
canada.dtypes
canada.columns = list(map(str,canada.columns));
```

### Viewing Line chart

```python
years = list(map(str,range(1980,2014)))
india_imgt = canada.loc['India', years]
india_imgt.plot();
india_imgt.plot(kind='line');
plt.title('Immigration from India');
plt.ylabel('Number of immigrants');
plt.xlabel('Years');
```

#### Plotting two countries and comparing

```python
canada.loc[['China','India'], years].transpose().plot();
plt.title('Immigration of China and India');
plt.xlabel('Years');
plt.ylabel('Immigrants');
```

#### As it is impossible to compare 195 countries and their charts, Last 3 year's average is taken and compared in the following steps:

### 

### Taking avg. of last 3 years and cleaning the Series

```python
canada['avglast3'] = (canada['2011']+canada['2012']+canada['2013'])/3
l3 = canada['avglast3']
l3 = l3.sort_values(ascending = False )
l3 = l3.round()
l3 = l3[l3>2]
```

```python
l3.head(10)
```

### Removing outliers

```python
l5 = l3[l3<11000];
```

#### Calculating Standard Deviation for bins

```python
np.std(l5)
```

### Creating bins

```python
type(l5)
dl5 = df(l5)
dl5
ahg = pd.Series(range(0,12000,1630))
bins = ahg
dl5['binned'] = pd.cut(dl5['avglast3'], bins)
```

### Plotting the bar graph

```python
out = dl5['binned']
ax = out.value_counts(sort=False).plot.bar(rot=0, color="b", figsize=(16,4))
plt.show()
```

### Bin values

```python
dl5['binned'].value_counts(sort=True)
```
